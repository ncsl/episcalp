```python
def calculate_distribution(arr, normalize=True):
    row_sum = np.sum(arr, axis=1)
    if not normalize:
        return row_sum
    row_sum_pdf = row_sum / np.sum(row_sum)
    return row_sum_pdf
```

```python
def create_normal_distribution(n_chs):
    dist = np.ones((n_chs, 1))/n_chs
    return dist
```

```python
def calculate_entropy(dist):
    return entropy(dist)
```

```python
def calculate_variance(dist):
    return np.var(dist)
```

```python
def calculate_skew(dist):
    return skew(dist)
```

```python
def calculate_kurtosis(dist):
    return kurtosis(dist)
```

```python
def calculate_kl_div(dist):
    comparison_dist = create_normal_distribution(len(dist))
    p = dist
    q = np.array([c[0] for c in comparison_dist]).reshape(-1, 1)
    print(f"p: {p.shape}, q: {q.shape}")
    return entropy(p, q)
```

```python
for subject in subjects:
    deriv_path = deriv_root / deriv_chain / f"sub-{subject}"
    sourcesink_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-sourcesink_ieeg.json"
    try:
        ss_ind_deriv = DerivativeNumpy(sourcesink_fpath)
    except FileNotFoundError:
        sourcesink_fpath = Path(sourcesink_fpath.parent) / sourcesink_fpath.name.replace("initialvisit", "awake")
        ss_ind_deriv = DerivativeNumpy(sourcesink_fpath)
    ss_ind = ss_ind_deriv.get_data()
    ss_pmf = calculate_distribution(ss_ind, True).reshape(-1, 1)
    ss_pmf_deriv = DerivativeArray(ss_pmf, info=ss_ind_deriv.info)
    pmf_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-pmf_eeg.npy"
    ss_pmf_deriv.save(pmf_fpath, overwrite=True)
    kl_div = calculate_kl_div(ss_pmf)
    kl_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-kldiv_eeg.npy"
    np.save(kl_fpath, kl_div)
    entr = calculate_entropy(ss_pmf)
    entropy_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-entropy_eeg.npy"
    np.save(entropy_fpath, entr)
    var = calculate_variance(ss_pmf)
    variance_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-variance_eeg.npy"
    np.save(variance_fpath, var)
    skew_val = calculate_skew(ss_pmf)
    skew_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-skew_eeg.npy"
    np.save(skew_fpath, skew_val)
    kurtosis_val = calculate_kurtosis(ss_pmf)
    kurtosis_fpath = deriv_path / f"sub-{subject}_ses-{session}_task-{task}_run-{run}_desc-kurtosis_eeg.npy"
    np.save(kurtosis_fpath, kurtosis_val)
```
