```python
raw = mne.io.read_raw_persyst(source_fpath, preload=True)
annotations = raw.annotations
sfreq = raw.info["sfreq"]
```

```python
raw.filter(0.5, 30)
raw.drop_channels(raw.info['bads'])
```

```python
spike_channels = [
    a["description"].split(" ")[1]
    for a in annotations
    if "spike" in a["description"].lower()
]
spike_channels = [s.split("-")[0] for s in spike_channels]
spike_onsets = [
    a["onset"]
    for a in annotations
    if "spike" in a["description"].lower()
]
for sc, so in zip(spike_channels, spike_onsets):
    print(f"{sc}: {so}")
```
